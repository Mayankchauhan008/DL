```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
# Replace "your_image_name.jpg" with the name of the image file you upload to Colab
image = cv2.imread("386745.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Define a vertical edge detection kernel (Sobel-like)
vertical_kernel = np.array([
    [+1, 0, -1],
    [+2, 0, -2],
    [+1, 0, -1]
])

# Apply convolution using filter2D
vertical_edges = cv2.filter2D(gray, -1, vertical_kernel)

# Threshold to make it binary: vertical lines = white, rest = black
_, binary_output = cv2.threshold(vertical_edges, 50, 255, cv2.THRESH_BINARY)

# Show original and edge-detected images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Grayscale Image')
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Vertical Edges Detected')
plt.imshow(binary_output, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
